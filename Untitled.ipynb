```python
date_columns = ['5/31/2023', '6/30/2023', '7/31/2023', '8/31/2023', '9/30/2023', '10/31/2023', 
                '11/30/2023', '12/31/2023', '1/31/2024', '2/29/2024', '3/31/2024', '4/30/2024', 
                '5/31/2024', '6/30/2024', '7/31/2024']

output = widgets.Output()

def display_model_accuracies(change=None):
    with output:
        output.clear_output()  # Clear previous outputs

        # Initializing lists to store accuracy metrics for each date
        r2_xgb_list, mae_xgb_list, rmse_xgb_list = [], [], []
        r2_rf_list, mae_rf_list, rmse_rf_list = [], [], []
        r2_dt_list, mae_dt_list, rmse_dt_list = [], [], []

        # Loop through each date and calculate accuracy metrics
        for date in date_columns:
            features = df.drop(columns=['RegionID', 'RegionName', 'StateName', date])
            target = df[date]
            X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
            
            # XGBoost Regressor
            xgb_model = XGBRegressor(n_estimators=100, random_state=42)
            xgb_model.fit(X_train, y_train)
            y_pred_xgb = xgb_model.predict(X_test)
            
            # Random Forest Regressor
            rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
            rf_model.fit(X_train, y_train)
            y_pred_rf = rf_model.predict(X_test)
            
            # Decision Tree Regressor
            dt_model = DecisionTreeRegressor(random_state=42)
            dt_model.fit(X_train, y_train)
            y_pred_dt = dt_model.predict(X_test)
            
            # Collecting accuracy metrics for XGBoost
            r2_xgb_list.append(r2_score(y_test, y_pred_xgb))
            mae_xgb_list.append(mean_absolute_error(y_test, y_pred_xgb))
            rmse_xgb_list.append(np.sqrt(mean_squared_error(y_test, y_pred_xgb)))
            
            # Collecting accuracy metrics for RandomForest
            r2_rf_list.append(r2_score(y_test, y_pred_rf))
            mae_rf_list.append(mean_absolute_error(y_test, y_pred_rf))
            rmse_rf_list.append(np.sqrt(mean_squared_error(y_test, y_pred_rf)))
            
            # Collecting accuracy metrics for DecisionTree
            r2_dt_list.append(r2_score(y_test, y_pred_dt))
            mae_dt_list.append(mean_absolute_error(y_test, y_pred_dt))
            rmse_dt_list.append(np.sqrt(mean_squared_error(y_test, y_pred_dt)))

        # Calculating average accuracy metrics across all dates
        print("Average accuracy metrics across all dates:")

        # XGBoost metrics
        print(f"\nXGBoost:")
        print(f"Average R-squared (R²): {np.mean(r2_xgb_list):.4f}")
        print(f"Average Mean Absolute Error (MAE): {np.mean(mae_xgb_list):.4f}")
        print(f"Average Root Mean Squared Error (RMSE): {np.mean(rmse_xgb_list):.4f}")

        # RandomForest metrics
        print(f"\nRandomForest:")
        print(f"Average R-squared (R²): {np.mean(r2_rf_list):.4f}")
        print(f"Average Mean Absolute Error (MAE): {np.mean(mae_rf_list):.4f}")
        print(f"Average Root Mean Squared Error (RMSE): {np.mean(rmse_rf_list):.4f}")

        # DecisionTree metrics
        print(f"\nDecisionTree:")
        print(f"Average R-squared (R²): {np.mean(r2_dt_list):.4f}")
        print(f"Average Mean Absolute Error (MAE): {np.mean(mae_dt_list):.4f}")
        print(f"Average Root Mean Squared Error (RMSE): {np.mean(rmse_dt_list):.4f}")

# Call the function once to display the accuracies for all dates
display_model_accuracies()

# Display the output
display(output)
```

```python
# Convert 'Date' column to datetime format
df_melted['Date'] = pd.to_datetime(df_melted['Date'], errors='coerce')

# Define a function to plot ARIMA results and check accuracy for a specific region, with value annotations
def plot_arima_with_accuracy(region_name, train_size=0.9):
    # Ensure train_size is between 0.1 and 1.0
    train_size = max(0.1, min(1.0, train_size))
    
    # Filter data for the selected region
    region_data = df_melted[df_melted['RegionName'] == region_name]
    
    # Ensure data is sorted by Date and set Date as index
    region_data = region_data.sort_values(by='Date')
    region_data.set_index('Date', inplace=True)
    
    # Resample the data by month if it's not already monthly
    ts_data = region_data['Value'].resample('M').mean().dropna()  # Resample to monthly and remove missing values
    
    # Split the data into train and test sets based on train_size
    train_size_idx = int(len(ts_data) * train_size)
    train_data = ts_data[:train_size_idx]
    test_data = ts_data[train_size_idx:]
    
    # Fit ARIMA model on the training data
    model = pm.auto_arima(train_data, seasonal=False, stepwise=True, suppress_warnings=True)
    
    # Forecast for the next 6 months
    forecast, conf_int = model.predict(n_periods=6, return_conf_int=True)
    
    # Handle zero or near-zero values in test_data to avoid NaN in MAPE
    mask = (test_data != 0)  # Only use data points where the actual values are not zero
    valid_test_data = test_data[mask]
    
    # Get in-sample predictions for the test period and align the forecast index with test data
    in_sample_forecast = model.predict_in_sample(start=len(train_data), end=len(ts_data)-1)
    valid_forecast = pd.Series(in_sample_forecast, index=test_data.index)[mask]
    
    # Calculate accuracy metrics
    mae = mean_absolute_error(valid_test_data, valid_forecast)
    rmse = np.sqrt(mean_squared_error(valid_test_data, valid_forecast))
    
    if not valid_test_data.empty:  # Only calculate MAPE if valid data exists
        mape = np.mean(np.abs((valid_test_data - valid_forecast) / valid_test_data)) * 100  # Mean Absolute Percentage Error
    else:
        mape = None
    
    print(f'Accuracy Metrics for {region_name}:')
    print(f'Mean Absolute Error (MAE): {mae:.4f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
    if mape is not None:
        print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
    else:
        print('Mean Absolute Percentage Error (MAPE): Not available (division by zero)')
    
    # Plot the original and forecasted data
    plt.figure(figsize=(15, 6))
    
    # Plot training data
    plt.plot(train_data, label='Training Data', marker='o')
    for i, value in enumerate(train_data):
        plt.text(train_data.index[i], value, f'{value:.2f}', fontsize=9, verticalalignment='bottom', horizontalalignment='right')
    
    # Plot test data
    plt.plot(test_data.index, test_data, label='Test Data', marker='o')
    for i, value in enumerate(test_data):
        plt.text(test_data.index[i], value, f'{value:.2f}', fontsize=9, verticalalignment='bottom', horizontalalignment='right')
    
    # Generate forecast index for the next 6 months
    forecast_index = pd.date_range(start=test_data.index[-1], periods=6, freq='M')
    
    # Plot forecasted data
    plt.plot(forecast_index, forecast, label='Forecasted Data (6 months)', color='orange', marker='o')
    for i, value in enumerate(forecast):
        plt.text(forecast_index[i], value, f'{value:.2f}', fontsize=9, verticalalignment='bottom', horizontalalignment='left', color='orange')
    
    # Plot confidence intervals
    plt.fill_between(forecast_index, conf_int[:, 0], conf_int[:, 1], color='orange', alpha=0.3)
    
    plt.title(f'ARIMA Forecast vs Actual for {region_name} (Next 6 Months)')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

    
    

# Get the unique list of region names
region_names = df_melted['RegionName'].unique()

# Use interact to create a dropdown list for region selection and accuracy check, including a slider for train_size
interact(plot_arima_with_accuracy, 
         region_name=widgets.Dropdown(value="New Haven",options=region_names, description='Region:'), 
         train_size=widgets.FloatSlider(value=0.9, min=0.1, max=1.0, step=0.05, description='Train Size'));
```

```python

# Define a function to plot SARIMA results and check accuracy for a specific region, allowing manual SARIMA parameters
def plot_sarima_with_accuracy(region_name, train_size=0.9, seasonal_period=6, p=1, d=1, q=1, P=1, D=1, Q=1):
    # Ensure train_size is between 0.1 and 1.0
    train_size = max(0.1, min(1.0, train_size))
    
    # Filter data for the selected region
    region_data = df_melted[df_melted['RegionName'] == region_name]
    
    # Print how many data points are available for this region
    print(f"Initial data points for {region_name}: {len(region_data)}")
    if region_data.empty:
        print(f"No data available for {region_name}.")
        return
    
    # Ensure data is sorted by Date and set Date as index
    region_data = region_data.sort_values(by='Date')
    region_data.set_index('Date', inplace=True)
    
    # Resample the data by month and drop missing values
    ts_data = region_data['Value'].resample('M').mean().dropna()  # Resample to monthly and remove missing values
    print(f"Data points after resampling for {region_name}: {len(ts_data)}")
    
    # Check if there's sufficient data after resampling
    if ts_data.empty or len(ts_data) < 2 * seasonal_period:
        print(f"Insufficient data for {region_name}. Please ensure there are enough data points.")
        return

    # Split the data into train and test sets based on train_size
    train_size_idx = int(len(ts_data) * train_size)
    train_data = ts_data[:train_size_idx]
    test_data = ts_data[train_size_idx:]

    # Fit SARIMA model on the training data
    try:
        model = pm.ARIMA(order=(p, d, q), seasonal_order=(P, D, Q, seasonal_period), suppress_warnings=True)
        model.fit(train_data)
    except ValueError as e:
        print(f"Error fitting SARIMA model: {e}")
        return
    
    # Forecast for the next 6 months
    forecast, conf_int = model.predict(n_periods=6, return_conf_int=True)
    
    # Handle zero or near-zero values in test_data to avoid NaN in MAPE
    mask = (test_data != 0)  # Only use data points where the actual values are not zero
    valid_test_data = test_data[mask]
    
    # Get in-sample predictions for the test period and align the forecast index with test data
    in_sample_forecast = model.predict_in_sample(start=len(train_data), end=len(ts_data)-1)
    valid_forecast = pd.Series(in_sample_forecast, index=test_data.index)[mask]
    
    # Calculate accuracy metrics
    mae = mean_absolute_error(valid_test_data, valid_forecast)
    rmse = np.sqrt(mean_squared_error(valid_test_data, valid_forecast))
    
    if not valid_test_data.empty:  # Only calculate MAPE if valid data exists
        mape = np.mean(np.abs((valid_test_data - valid_forecast) / valid_test_data)) * 100  # Mean Absolute Percentage Error
    else:
        mape = None
    
    print(f'Accuracy Metrics for {region_name}:')
    print(f'Mean Absolute Error (MAE): {mae:.4f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
    if mape is not None:
        print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
    else:
        print('Mean Absolute Percentage Error (MAPE): Not available (division by zero)')
    
    # Plot the original and forecasted data
    plt.figure(figsize=(15, 6))
    
    # Plot training data
    plt.plot(train_data, label='Training Data', marker='o')
    for i, value in enumerate(train_data):
        plt.text(train_data.index[i], value, f'{value:.2f}', fontsize=9, verticalalignment='bottom', horizontalalignment='right')
    
    # Plot test data
    plt.plot(test_data.index, test_data, label='Test Data', marker='o')
    for i, value in enumerate(test_data):
        plt.text(test_data.index[i], value, f'{value:.2f}', fontsize=9, verticalalignment='bottom', horizontalalignment='right')
    
    # Generate forecast index for the next 6 months
    forecast_index = pd.date_range(start=test_data.index[-1], periods=6, freq='M')
    
    # Plot forecasted data
    plt.plot(forecast_index, forecast, label='Forecasted Data (6 months)', color='orange', marker='o')
    for i, value in enumerate(forecast):
        plt.text(forecast_index[i], value, f'{value:.2f}', fontsize=9, verticalalignment='bottom', horizontalalignment='left', color='orange')
    
    # Plot confidence intervals
    plt.fill_between(forecast_index, conf_int[:, 0], conf_int[:, 1], color='orange', alpha=0.3)
    
    plt.title(f'SARIMA Forecast vs Actual for {region_name} (Next 6 Months)')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Get the unique list of region names
region_names = df_melted['RegionName'].unique()

# Use interact to create a dropdown list for region selection and accuracy check, including a slider for train_size, seasonal period, and SARIMA parameters
interact(plot_sarima_with_accuracy, 
         region_name=widgets.Dropdown(value="New Haven", options=region_names, description='Region:'), 
         train_size=widgets.FloatSlider(value=0.9, min=0.1, max=1.0, step=0.05, description='Train Size'),
         seasonal_period=widgets.IntSlider(value=6, min=1, max=12, step=1, description='Seasonal Period'),
         p=widgets.IntSlider(value=1, min=0, max=5, step=1, description='p'),
         d=widgets.IntSlider(value=1, min=0, max=2, step=1, description='d'),
         q=widgets.IntSlider(value=1, min=0, max=5, step=1, description='q'),
         P=widgets.IntSlider(value=1, min=0, max=5, step=1, description='P'),
         D=widgets.IntSlider(value=1, min=0, max=2, step=1, description='D'),
         Q=widgets.IntSlider(value=1, min=0, max=5, step=1, description='Q'));
```

```python
def prepare_lstm_data(data, n_lag):
    X, y = [], []
    for i in range(len(data) - n_lag):
        X.append(data[i:i + n_lag])
        y.append(data[i + n_lag])
    return np.array(X), np.array(y)

# Function to calculate MAPE (Mean Absolute Percentage Error)
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Function to forecast the next 6 months and calculate accuracy metrics
def plot_lstm_forecast(region_name, train_size=0.8, lstm_units=50, dropout_rate=0.2, epochs=50, batch_size=32, optimizer_choice='adam'):
    n_lag = 3  # Number of previous time steps used as input
    forecast_steps = 6  # Number of months to forecast

    # Filter data for the selected region
    region_data = df_melted[df_melted['RegionName'] == region_name]
    
    # Ensure data is sorted by Date and set Date as index
    region_data = region_data.sort_values(by='Date')
    region_data.set_index('Date', inplace=True)
    
    # Resample the data by month if it's not already monthly
    ts_data = region_data['Value'].resample('M').mean().dropna()  # Resample to monthly and remove missing values
    
    # Scale the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    ts_scaled = scaler.fit_transform(ts_data.values.reshape(-1, 1))
    
    # Split the data into train and test sets based on train_size
    train_size_idx = int(len(ts_scaled) * train_size)
    train_data = ts_scaled[:train_size_idx]
    
    # Prepare LSTM data
    X_train, y_train = prepare_lstm_data(train_data, n_lag)
    
    # Reshape input to be [samples, time steps, features]
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    
    # Build the LSTM model using Input layer for the first input shape
    model = Sequential()
    model.add(Input(shape=(X_train.shape[1], 1)))  # Use Input layer explicitly
    model.add(LSTM(lstm_units, return_sequences=True))
    model.add(Dropout(dropout_rate))  # Add dropout layer for regularization
    model.add(LSTM(lstm_units))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))
    
    # Select optimizer
    if optimizer_choice == 'adam':
        optimizer = Adam()
    elif optimizer_choice == 'rmsprop':
        optimizer = RMSprop()
    else:
        optimizer = 'adam'  # Default
    
    # Compile the model
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    
    # Train the model
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1, validation_split=0.1)
    
    # Forecast for the next 6 months
    last_data_points = ts_scaled[train_size_idx-n_lag:train_size_idx]  # Use last n_lag data points
    forecast_input = last_data_points.reshape(1, n_lag, 1)  # Reshape for LSTM input
    
    forecasted_values = []
    for _ in range(forecast_steps):
        # Make prediction for one step
        next_value = model.predict(forecast_input)
        forecasted_values.append(next_value[0, 0])  # Store the prediction
        
        # Reshape the next_value and update the input for the next step
        next_value_reshaped = next_value.reshape(1, 1, 1)  # Reshape to [samples, time steps, features]
        forecast_input = np.append(forecast_input[:, 1:, :], next_value_reshaped, axis=1)
    
    # Inverse scaling to get the forecasted values back to original scale
    forecasted_values = scaler.inverse_transform(np.array(forecasted_values).reshape(-1, 1))
    
    # Generate future dates for the forecasted values
    last_date = ts_data.index[train_size_idx-1]
    forecast_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=forecast_steps, freq='M')
    
    # Calculate accuracy metrics for the training data
    train_predict = model.predict(X_train)
    train_predict = scaler.inverse_transform(train_predict)
    y_train_rescaled = scaler.inverse_transform(y_train.reshape(-1, 1))
    
    mae = mean_absolute_error(y_train_rescaled, train_predict)
    rmse = np.sqrt(mean_squared_error(y_train_rescaled, train_predict))
    mape = mean_absolute_percentage_error(y_train_rescaled, train_predict)
    
    print(f'Accuracy Metrics for {region_name}:')
    print(f'Mean Absolute Error (MAE): {mae:.4f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
    print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
    
    # Plot the original data and the forecasted values
    plt.figure(figsize=(15, 6))
    
    # Plot historical data
    plt.plot(ts_data.index[:train_size_idx], ts_data.values[:train_size_idx], label='Historical Data', marker='o')
    
    # Annotate values on historical data
    for i, value in enumerate(ts_data.values[:train_size_idx]):
        plt.text(ts_data.index[i], value, f'{value:.2f}', fontsize=8, verticalalignment='bottom', horizontalalignment='right')
    
    # Plot forecasted data
    plt.plot(forecast_dates, forecasted_values, label='Forecasted Data (Next 6 Months)', color='orange', marker='o')
    
    # Annotate values on forecasted data
    for i, value in enumerate(forecasted_values):
        plt.text(forecast_dates[i], value[0], f'{value[0]:.2f}', fontsize=8, verticalalignment='bottom', horizontalalignment='right')
    
    plt.title(f'LSTM Forecast for the Next 6 Months for {region_name}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Get the unique list of region names
region_names = df_melted['RegionName'].unique()

# Use interact to create a dropdown list for region selection and hyperparameter sliders
interact(plot_lstm_forecast, 
         region_name=widgets.Dropdown(value="New Haven",options=region_names, description='Region:'),
         train_size=widgets.FloatSlider(value=0.9, min=0.1, max=0.9, step=0.05, description='Train Size'),
         lstm_units=widgets.IntSlider(value=50, min=10, max=200, step=10, description='LSTM Units'),
         dropout_rate=widgets.FloatSlider(value=0.15, min=0.0, max=0.5, step=0.05, description='Dropout Rate'),
         epochs=widgets.IntSlider(value=50, min=10, max=200, step=10, description='Epochs'),
         batch_size=widgets.IntSlider(value=322, min=8, max=128, step=8, description='Batch Size'),
         optimizer_choice=widgets.Dropdown(options=['adam', 'rmsprop'], description='Optimizer'));
```

```python

# Function to create lagged features for time series
def create_lagged_features(data, n_lag=3):
    df = pd.DataFrame(data)
    for i in range(1, n_lag+1):
        df[f'lag_{i}'] = df[0].shift(i)
    df.dropna(inplace=True)  # Drop rows with missing values
    return df

# Function to calculate MAPE (Mean Absolute Percentage Error)
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Function to forecast the next 6 months using XGBoost
def plot_xgboost_forecast(region_name, train_size=0.8, n_lag=3):
    forecast_steps = 6  # Number of months to forecast

    # Filter data for the selected region
    region_data = df_melted[df_melted['RegionName'] == region_name]
    
    # Ensure data is sorted by Date and set Date as index
    region_data = region_data.sort_values(by='Date')
    region_data.set_index('Date', inplace=True)
    
    # Resample the data by month if it's not already monthly
    ts_data = region_data['Value'].resample('M').mean().dropna()  # Resample to monthly and remove missing values
    
    # Create lagged features
    ts_lagged = create_lagged_features(ts_data.values, n_lag=n_lag)
    
    # Split data into features (X) and target (y)
    X = ts_lagged.drop(0, axis=1).values  # Lagged features
    y = ts_lagged[0].values  # Target
    
    # Split the data into training and test sets based on train_size
    train_size_idx = int(len(X) * train_size)
    X_train, X_test, y_train, y_test = X[:train_size_idx], X[train_size_idx:], y[:train_size_idx], y[train_size_idx:]
    
    # Train the XGBoost model
    model = XGBRegressor(objective='reg:squarederror', n_estimators=100)
    model.fit(X_train, y_train)
    
    # Forecast the next 6 months
    last_data_points = X[-n_lag:]  # Use the last `n_lag` data points
    forecasted_values = []
    
    current_input = last_data_points[-1].reshape(1, -1)
    
    for _ in range(forecast_steps):
        next_value = model.predict(current_input)[0]
        forecasted_values.append(next_value)
        
        # Create new input with the predicted value and update lagged features
        current_input = np.roll(current_input, -1)
        current_input[0, -1] = next_value
    
    # Generate future dates for the forecasted values
    last_date = ts_data.index[-1]
    forecast_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=forecast_steps, freq='M')
    
    # Calculate accuracy metrics for the test data
    y_pred = model.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape = mean_absolute_percentage_error(y_test, y_pred)
    
    print(f'Accuracy Metrics for {region_name}:')
    print(f'Mean Absolute Error (MAE): {mae:.4f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
    print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
    
    # Plot the original data and the forecasted values
    plt.figure(figsize=(15, 6))
    
    # Plot historical data
    plt.plot(ts_data.index[:train_size_idx], ts_data.values[:train_size_idx], label='Historical Data', marker='o')
    
    # Plot forecasted data
    plt.plot(forecast_dates, forecasted_values, label='Forecasted Data (Next 6 Months)', color='orange', marker='o')
    
    # Annotate values on forecasted data
    for i, value in enumerate(forecasted_values):
        plt.text(forecast_dates[i], value, f'{value:.2f}', fontsize=8, verticalalignment='bottom', horizontalalignment='right')
    
    plt.title(f'XGBoost Forecast for the Next 6 Months for {region_name}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Get the unique list of region names
region_names = df_melted['RegionName'].unique()

# Use interact to create a dropdown list for region selection and a slider for train_size and lag
interact(plot_xgboost_forecast, 
         region_name=widgets.Dropdown(value="New Haven",options=region_names, description='Region:'),
         train_size=widgets.FloatSlider(value=0.95, min=0.1, max=1.0, step=0.05, description='Train Size'),
         n_lag=widgets.IntSlider(value=5, min=1, max=12, step=1, description='Lag (Months)'));
```

```python
def plot_prophet_with_accuracy(region_name, train_size=0.9):
    train_size = max(0.1, min(1.0, train_size))
    
    region_data = df_melted[df_melted['RegionName'] == region_name]
    region_data = region_data.sort_values(by='Date')
    region_data.set_index('Date', inplace=True)
    
    ts_data = region_data['Value'].resample('M').mean().dropna()
    
    train_size_idx = int(len(ts_data) * train_size)
    train_data = ts_data[:train_size_idx]
    test_data = ts_data[train_size_idx:]
    
    # Apply log transformation to stabilize variance
    prophet_data = train_data.reset_index().rename(columns={"Date": "ds", "Value": "y"})
    prophet_data['y'] = np.log1p(prophet_data['y'])
    
    # Initialize and fit the Prophet model with higher changepoint sensitivity and yearly seasonality
    model = Prophet(
        seasonality_mode='multiplicative',
        changepoint_prior_scale=0.5,  # Increase sensitivity to changepoints
        seasonality_prior_scale=0.1,
        interval_width=0.95
    )
    model.add_seasonality(name='monthly', period=30.5, fourier_order=5)
    model.add_seasonality(name='yearly', period=365.25, fourier_order=10)  # Add yearly seasonality
    model.fit(prophet_data)
    
    # Create future dates and make predictions for the next 6 months
    future = model.make_future_dataframe(periods=6, freq='M')
    forecast = model.predict(future)
    
    # Invert the log transformation
    forecast['yhat'] = np.expm1(forecast['yhat'])
    forecast['yhat_lower'] = np.expm1(forecast['yhat_lower'])
    forecast['yhat_upper'] = np.expm1(forecast['yhat_upper'])
    
    # Filter only future forecast data
    future_forecast = forecast[forecast['ds'] > train_data.index[-1]]
    
    # Calculate accuracy metrics on the test data
    forecasted_values = forecast.set_index('ds').loc[test_data.index, 'yhat']
    mae = mean_absolute_error(test_data, forecasted_values)
    rmse = np.sqrt(mean_squared_error(test_data, forecasted_values))
    mape = np.mean(np.abs((test_data - forecasted_values) / test_data)) * 100 if not test_data.empty else None
    
    print(f'Accuracy Metrics for {region_name}:')
    print(f'MAE: {mae:.4f}' if mae else 'MAE: Not available')
    print(f'RMSE: {rmse:.4f}' if rmse else 'RMSE: Not available')
    print(f'MAPE: {mape:.2f}%' if mape else 'MAPE: Not available')
    
    # Plotting
    plt.figure(figsize=(15, 6))
    plt.plot(train_data, label='Training Data', marker='o')
    plt.plot(test_data, label='Test Data', marker='o')
    plt.plot(future_forecast['ds'], future_forecast['yhat'], label='Forecasted Data (Next 6 Months)', color='orange', marker='o')
    plt.fill_between(future_forecast['ds'], future_forecast['yhat_lower'], future_forecast['yhat_upper'], color='orange', alpha=0.3)
    
    plt.title(f'Prophet Forecast vs Actual for {region_name} (Next 6 Months)')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Use interact to select region and adjust train size
interact(plot_prophet_with_accuracy, 
         region_name=widgets.Dropdown(value="New Haven", options=region_names, description='Region:'), 
         train_size=widgets.FloatSlider(value=0.9, min=0.1, max=1.0, step=0.05, description='Train Size'));
```

```python

import traceback

# Create an empty DataFrame to store the forecasts
df_forecast = pd.DataFrame(columns=['RegionID', 'RegionName', 'StateName', 'Date', 'Value'])

# Define a function to forecast and store results for all regions
def generate_forecasts_for_all_regions(train_size=0.9):
    global df_forecast  # Use the global variable to accumulate forecasts
    
    # Ensure train_size is between 0.1 and 1.0
    train_size = max(0.1, min(1.0, train_size))
    
    # Loop through each unique region in df_melted
    for region_name in df_melted['RegionName'].unique():
        try:
            # Debug information
            print(f"Processing region: {region_name}")
            
            # Filter data for the current region
            region_data = df_melted[df_melted['RegionName'] == region_name]
            
            # Ensure data is sorted by Date and set Date as index
            region_data = region_data.sort_values(by='Date')
            region_data.set_index('Date', inplace=True)
            
            # Resample the data by month if it's not already monthly
            ts_data = region_data['Value'].resample('M').mean().dropna()
            
            # Check if there's sufficient data for ARIMA modeling
            if len(ts_data) < 12:
                print(f"Skipping region {region_name} due to insufficient data.")
                continue
            
            # Split the data into train and test sets based on train_size
            train_size_idx = int(len(ts_data) * train_size)
            train_data = ts_data[:train_size_idx]
            
            # Fit ARIMA model on the training data
            model = pm.auto_arima(
                train_data, 
                start_p=0, start_q=0, max_p=2, max_q=2,
                seasonal=False, stepwise=True, 
                suppress_warnings=True,
                error_action='ignore', trace=True
            )
            
            # Forecast for the next 6 months
            forecast, conf_int = model.predict(n_periods=6, return_conf_int=True)
            
            # Generate forecast index starting from the end of the entire dataset (not just training)
            forecast_start_date = ts_data.index[-1] + pd.DateOffset(months=1)
            forecast_index = pd.date_range(start=forecast_start_date, periods=6, freq='M')
            
            # Get region details for the new DataFrame
            region_id = region_data['RegionID'].iloc[0]
            state_name = region_data['StateName'].iloc[0]
            
            # Create a DataFrame with forecasted values
            forecast_df = pd.DataFrame({
                'RegionID': region_id,
                'RegionName': region_name,
                'StateName': state_name,
                'Date': forecast_index,
                'Value': forecast
            })
            
            # Drop any empty or all-NA columns before concatenation
            forecast_df = forecast_df.dropna(how='all', axis=1)
            
            # Append the forecasted data to df_forecast
            df_forecast = pd.concat([df_forecast, forecast_df], ignore_index=True)
        
        except Exception as e:
            # Print detailed error message with traceback
            print(f"Error processing region {region_name}:")
            traceback.print_exc()

# Run the function to generate forecasts for all regions
generate_forecasts_for_all_regions(train_size=0.9)

# Display the first few rows of df_forecast to verify
print(df_forecast.head())
```
