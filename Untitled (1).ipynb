```python
import pandas as pd 
import geopy
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv("C:\\Users\\chand\\Downloads\\df_melted.csv",header=None)
df = df[0].str.split(',', expand=True)
df.columns = ['RegionID', 'RegionName', 'StateName', 'Date', 'Value']
df.head()
```

```python
def get_lat_long(RegionName, StateName):
    address = f"{RegionName}, {StateName}"
    try:
        geolocator = Nominatim(user_agent="running", timeout=15)
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None
    
```

```python
df['LATITUDE'], df['LONGITUDE'] = zip(*df.apply(lambda x: get_lat_long(x['RegionName'], x['StateName']), axis=1))
```

```python
df['LATLONG'] = df['LATITUDE'].astype(str)+','+ df['LONGITUDE'].astype(str)
```
